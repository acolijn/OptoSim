```python
    def fresnel_coefficients(n1, n2, theta_i):
        """
        Calculate the reflection and transmission coefficients using the Fresnel equations.

        Parameters
        ----------
        n1(float) : float
            The refractive index of the medium in which the photon is incident
        n2(float) : float
            The refractive index of the medium in which the photon is transmitted
        theta_i(float) : float  
            The angle of incidence

        Returns
        -------
        r_parallel(float) : float
            The reflection coefficient for p-polarization
        t_parallel(float) : float
            The transmission coefficient for p-polarization
        r_perpendicular(float) : float
            The reflection coefficient for s-polarization
        t_perpendicular(float) : float  
            The transmission coefficient for s-polarization

        A.P. Colijn
        """
        # Calculate angles of refraction (theta_t) using Snell's law
        sin_theta_i = np.sin(theta_i)
        sin_theta_t = (n1 / n2) * sin_theta_i

        # Handle total internal reflection
        if sin_theta_t > 1.0:
            return 1.0, 0.0, 1.0, 0.0  # Total internal reflection, R=1, T=0

        cos_theta_i = np.cos(theta_i)
        cos_theta_t = np.sqrt(1.0 - sin_theta_t**2)

        print('cos_theta_i', cos_theta_i, 'cos_theta_t', cos_theta_t,'n1',n1,'n2',n2)

        # Calculate the coefficients for p-polarization
        r_parallel = ((n1 * cos_theta_i - n2 * cos_theta_t) /
                    (n1 * cos_theta_i + n2 * cos_theta_t))
        t_parallel = (2.0 * n1 * cos_theta_i) / (n1 * cos_theta_i + n2 * cos_theta_t)

        # Calculate the coefficients for s-polarization
        r_perpendicular = ((n2 * cos_theta_i - n1 * cos_theta_t) /
                        (n2 * cos_theta_i + n1 * cos_theta_t))
        t_perpendicular = (2.0 * n1 * cos_theta_i) / (n2 * cos_theta_i + n1 * cos_theta_t)
        
        f = n2*cos_theta_t/n1*cos_theta_i
        
        #print('theta_i', theta_i, 'theta_t', np.arcsin(sin_theta_t))
        #print("r_parallel", r_parallel**2, "t_parallel", t_parallel**2, "r_perpendicular", r_perpendicular**2, "t_perpendicular", t_perpendicular**2)
        return r_parallel**2, f*t_parallel**2, r_perpendicular**2, f*t_perpendicular
```

```python
rrs = []
tts = []
rrp = []
ttp = []
for th in theta:
    rs,ts, rp,tp= fresnel_coefficients(1.5,1.0,th)
    
    rrs.append(rs)
    tts.append(ts)
    rrp.append(rp)
    ttp.append(tp)
    
    
    
rrs = np.array(rrs)
tts = np.array(tts)
rrp = np.array(rrp)
ttp = np.array(ttp)

plt.plot(theta*180/3.1415,rrs,'blue')
plt.plot(theta*180/3.1415,rrp,linestyle='--',color='blue')

plt.plot(theta*180/3.1415,tts,'red')
plt.plot(theta*180/3.1415,ttp,linestyle='--',color='red')

plt.plot(theta*180/3.1415,tts+rrs,'green')
plt.plot(theta*180/3.1415,ttp+rrp,linestyle='--',color='green')
```

```python
import numpy as np

def fresnel_coefficients(n1, n2, theta_i):
    cos_i = np.cos(theta_i)
    sin_i = np.sin(theta_i)
    n_ratio = n1 / n2
    sin_t = n_ratio * sin_i
    cos_t = np.sqrt(1.0 - sin_t**2)
    
    rs = (n1 * cos_i - n2 * cos_t) / (n1 * cos_i + n2 * cos_t)
    rp = (n2 * cos_i - n1 * cos_t) / (n2 * cos_i + n1 * cos_t)
    
    R = (np.abs(rs)**2 + np.abs(rp)**2) / 2.0
    T = 1.0 - R
    
    return R, T

# Example usage:
n1 = 1.0  # Refractive index of the incident medium
n2 = 1.5  # Refractive index of the refractive medium
theta_i = np.radians(30)  # Incident angle in radians (e.g., 30 degrees)

R, T = fresnel_coefficients(n1, n2, theta_i)

# Verify that R + T equals 1 for energy conservation
total_energy = R + T

print("Reflection Coefficient:", R)
print("Transmission Coefficient:", T)
print("Total Energy (R + T):", total_energy)
```
