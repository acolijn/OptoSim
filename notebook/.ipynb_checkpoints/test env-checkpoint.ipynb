```python
import sys
sys.path.insert(0,'../python/')
from OpticalPhoton import *
import matplotlib.pyplot as plt
```

```python
myPhoton = OpticalPhoton()

xb = []
yb = []
xt = []
yt = []

for i in range(10):
    if i%10000 == 0:
        print('processed ',i,' photons')
    myPhoton.generate_photon(x0=(2.,0,0.1))
    myPhoton.propagate()
    
    if myPhoton.is_detected():
        x = myPhoton.get_photon_position()
        #myPhoton.print()
        if x[2]<0:
            xb.append(x[0])
            yb.append(x[1])
        else:
            xt.append(x[0])
            yt.append(x[1])
xb = np.array(xb)
yb = np.array(yb)
xt = np.array(xt)
yt = np.array(yt)

print(len(xb), len(xt))
```
