```python
from optosim.simulation.optical_photon_display import *
from optosim.settings import *
import matplotlib.pyplot as plt
```

#### Test the Fresnel coefficient implementation in the OpticalPhoton class

```python
config_file = CONFIG_DIR + '/config_nev10k_nph1k-25k.json'

myPhoton = OpticalPhotonDisplay(config=config_file)


# refractive indices
n1 = 1.6
n2 = 3.5
# theta range from 0-pi/2
theta = np.arange(0,np.pi/2,0.01)

R = np.empty_like(theta)
rs = np.empty_like(theta)
rp = np.empty_like(theta)
for i,th in enumerate(theta):
    R[i], _ = myPhoton.fresnel_coefficients_average(n1, n2, th)
    rs[i],rp[i] = myPhoton.fresnel_coefficients(n1,n2,th)

plt.plot(theta*180/3.1415,R,'blue')
plt.plot(theta*180/3.1415,rs**2,linestyle='-.',color='blue')
plt.plot(theta*180/3.1415,rp**2,linestyle='--',color='blue')

plt.plot(theta*180/3.1415,1-R,'red')
plt.plot(theta*180/3.1415,1-rs**2,linestyle='-.',color='red')
plt.plot(theta*180/3.1415,1-rp**2,linestyle='--',color='red')

plt.xlabel('$\\theta$')
plt.show()
```
